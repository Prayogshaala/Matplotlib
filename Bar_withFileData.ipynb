```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('C:\My_Folders\AI\Pandas\stack-overflow-developer-survey-2019\survey_results_public.csv')

Languages=df['LanguageWorkedWith']

from collections import Counter

Languages = Languages.fillna('').astype(str)

language_counter = Counter()

for response in Languages:
    if response:
        language_counter.update(response.split(';'))

print(language_counter)
```

```python
df=pd.read_csv('C:\My_Folders\AI\Pandas\stack-overflow-developer-survey-2019\survey_results_public.csv')

Languages=df['LanguageWorkedWith']

Languages = Languages.fillna('').astype(str)

language_cnt=Languages.str.split(';').explode().value_counts()

lang=language_cnt.index
count=language_cnt.values
plt.style.use('ggplot')
plt.barh(lang[::-1],count[::-1])
plt.ylabel('Languages')
plt.xlabel('Number of user')
plt.title('Users by programming languages')
plt.yticks(rotation=45)
plt.show()
```
