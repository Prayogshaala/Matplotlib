```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Data.csv')
Age=df['Age']
All_dev=df['All_Devs']
Py_dev=df['Python']
Java_dev=df['JavaScript']
```

```python
fig,(ax1,ax2)=plt.subplots(nrows=2,ncols=1,sharex=True)

ax1.plot(Age,All_dev,color='b',label='Mediam salary of All developers by age')
ax1.set_ylabel('Median Salary')
ax1.legend()

ax2.plot(Age,Py_dev,color='r',label='Mediam salary of Python developers by age')
ax2.set_xlabel('Age')
ax2.set_ylabel('Median Salary')
ax2.legend()

```
