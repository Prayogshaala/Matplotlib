```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
day=np.arange(1,11,1)

player1=np.random.randint(1,10,10)
player2=np.random.randint(1,10,10)
print(player1)
print(player2)
labels=['Player1','Player2']
```

```python
plt.stackplot(day,player1,player2,labels=labels)
plt.legend()
plt.show()
```
