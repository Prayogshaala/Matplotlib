```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('Data.csv')
df.head()

Age=df['Age']
All_devs=df['All_Devs']
Py_devs=df['Python']
```

```python
plt.scatter(Age,All_devs, c=All_devs,cmap='Greens',edgecolor='black') 
    #c=All_devs makes the marker color shade a function of values in All_devs list.
    #cmap='Greens' is a predefined style of coloring. There are other options available
    #edgecolor= Sets the color for the edge around the marker

cbar=plt.colorbar()
    # Creates a bar which would show what the shades of the color signify
cbar.set_label('Median Salary')
    # Sets the label for the color bar
plt.title('Median Salary by Age')
plt.xlabel('Age')
plt.ylabel('Median Salary')

plt.show()
```
