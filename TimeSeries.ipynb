```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdate
import numpy as np
```

```python
dates = np.arange(np.datetime64('2024-01-01'), np.datetime64('2024-01-11'))

Y=np.random.randint(1,10,10)
Y
```

```python
plt.plot_date(dates,Y,linestyle='solid')
plt.gcf().autofmt_xdate()  #Autoformat the dates displayed along x axis, rotates them at an angle
plt.gca().xaxis.set_major_formatter(mdate.DateFormatter('%Y-%m-%d')) # Sets the date format along x axis
plt.show
```
