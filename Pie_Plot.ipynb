```python
import matplotlib.pyplot as plt
```

```python
data=[45,30,17,8]
color=['red','blue','yellow','black']
labels=['One','Two','Three','Four']
explode=[0,0.2,0,0]
```

```python
plt.pie(data,colors=color,labels=labels,wedgeprops={'edgecolor':'black'}, explode=explode, shadow=True)
plt.show
```
