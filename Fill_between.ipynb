```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv('data.csv')
df.head()
```

```python
Age=df['Age']
All_dev=df['All_Devs']
Python_dev=df['Python']
```

```python
plt.plot(Age,All_dev,color='b',label='Median Salary for all devs')

plt.plot(Age,Python_dev,color='r',label='Median Salary for Python devs')

Threshold=55000

plt.fill_between(Age,Python_dev,Threshold, where=(Python_dev> Threshold),interpolate=True,alpha=0.25 )


plt.fill_between(Age,Python_dev,Threshold, where=(Python_dev<= Threshold),interpolate=True, color='y' )

plt.title('Median salary of developers by age')
plt.legend()
plt.style.use('classic')
plt.show()
```
